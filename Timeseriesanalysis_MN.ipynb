# Groupby data with condition and Rolling mean for detecting Variations

```python
data['TBD'] =data.TBD.convert_objects(convert_numeric=True)
```

```python
Chl_TBD = data.where((data["TBD"] >0.5)).dropna()[["Chlorine_amount", "TBD"]]
Chlo_TBD = Chl_TBD.groupby(Chl_TBD.index).mean()
```

```python
Chlo_TBD.info()
```

```python
fig = plt.figure(figsize=(20,10))
ax=fig.add_subplot(111)
ax.set_ylim([0,1])
ax.set_ylabel("Total solid", fontsize=20)
ax.set_xlabel("Time", fontsize=25)
ax.yaxis.label.set_color("red")

ax2=ax.twinx()
ax2.set_ylim([0.5,2])
ax2.set_ylabel("Chlorine amount( mg/L)", fontsize = 20)

ax2.yaxis.label.set_color("green")

Chlo_TBD["Chlorine_amount"].rolling(12).mean().plot(color="red", fontsize =20)
Chlo_TBD["TBD"].rolling(12).mean().plot(color ="green", fontsize=20)

plt.grid(False)
```
